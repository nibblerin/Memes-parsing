```python
def getMemesPage(page):
  page_link = 'https://knowyourmeme.com/memes/page/{}'.format(page)
  response = get(page_link, headers={'User-Agent': UserAgent().chrome})
  if not response.ok:
    return []
  html = response.content
  soup = BeautifulSoup(html, 'html.parser')
  meme_links = soup.find_all('a', class_=['item', 'has-quick-info'])
  meme_links = ['https://knowyourmeme.com/memes' + link.get('href') for link in meme_links]
  return meme_links
```

```python
def getProperties(soup):
  name_meme = soup.find('section',{'class':'info'}).h1.text.strip()
  properties = soup.find('aside', {'class': 'left'})

  status_meme = properties.find('dd')
  status_meme = '' if not status_meme else status_meme.text.strip()

  type_meme = properties.find('a', {'class': 'entry-type-link'})
  type_meme = '' if not type_meme else type_meme.text

  year_origin_meme = properties.find('dt', string='\nYear\n')
  year_origin_meme = '' if not year_origin_meme else year_origin_meme.find_next_sibling('dd')
  year_origin_meme = year_origin_meme.text.strip()

  place_origin_meme = properties.find('dd', {'class': 'entry_origin_link'})
  place_origin_meme = '' if not place_origin_meme else place_origin_meme.text.strip()

  tags_meme = properties.find('dl', {'id': 'entry_tags'}).find_all('a')
  tags_meme = '' if not tags_meme else ", ".join(a.text.strip() for a in tags_meme)

  return name_meme, status_meme, type_meme, year_origin_meme, place_origin_meme, tags_meme
```

```python
from os.path import abspath
def getText(soup):
  body = soup.find('section',{'class':'bodycopy'})

  about_meme = body.find('p')
  about_meme = '' if not about_meme else about_meme.text

  about_meme = body.find('h2', {'id': 'about'})
  about_text = []
  for about in about_meme.find_next_siblings():
      if about.name == "h2":
          break
      if about.name == "p":
          about_text.append(about.get_text(strip=True))
  about_meme = ' '.join(about_text).strip()

  origin_meme = body.find('h2', {'id': 'origin'})
  origin_text = []
  for origin in origin_meme.find_next_siblings():
      if origin.name == "h2":
          break
      elif origin.name == "center":
          continue
      else:
          origin_text.append(origin.get_text(strip=True))
  origin_meme = ' '.join(origin_text).strip()

  skip_ids = {"about", "origin"}
  collect = False
  other_text_list = []

  for el in body.children:
      if el.name == "h2":
          if el.get("id") in skip_ids:
              collect = False
          else:
              collect = True
          continue

      if collect and el.name == "p":
          other_text_list.append(el.get_text(strip=True))

  other_text = ' '.join(other_text_list).strip()

  return about_meme, origin_meme, other_text
```
